```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Año de la NBA que analizaremos
year = 2020
#year = range(2018, 2021)
#year_list = list(year)

#URL de la página a scrapear
url = 'https://www.basketball-reference.com/leagues/NBA_{}_totals.html'.format(year)

#HTML de la URL
html = urlopen(url)

soup = BeautifulSoup(html)
```

```python
#Podemos utilizar findAll() para conseguir la columna de encabezados
soup.findAll('tr', limit = 1)


#Utilizamos getText() para extraer lo que necesitamos en una lista
headers_key = [th.getText() for th in soup.findAll('tr', limit=1)[0].findAll('th')]
headers_value = [th.get('aria-label') for th in soup.findAll('tr', limit=1)[0].findAll('th')]

#Excluimos la primera columna
headers_key = headers_key[1:]
headers_value = headers_value[1:]

#Creamos un diccionario con el significado de la cabecera de las tablas
i = 0
headers_dict = {}

for k in headers_key:
    headers_dict[headers_key[i]] = headers_value[i]
    i+=1

headers_dict
```

```python
#El siguiente paso es extraer los datos de las celdas de la tabla con las estadísticas. 
#Evitamos la fila de encabezado
rows = soup.findAll('tr')[1:]
player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
```

```python
#Ahora que tenemos los datos y las cabeceras, creamos nuestro dataframe. 
stats = pd.DataFrame(player_stats, columns = headers_key)
stats.head(10)
```

```python
#Observando la web vemos que cada 20 filas aparece de nuevo los encabezados, filtramos esas fila de nuestro dataset
stats.head(30)
```
